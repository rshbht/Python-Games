```python
import random

# initialize the board
board = [" " for i in range(9)]

# define a function to print the board
def print_board():
    print("-------------")
    print("| {} | {} | {} |".format(board[0], board[1], board[2]))
    print("-------------")
    print("| {} | {} | {} |".format(board[3], board[4], board[5]))
    print("-------------")
    print("| {} | {} | {} |".format(board[6], board[7], board[8]))
    print("-------------")

# define a function to check if a player has won
def check_win(player):
    return ((board[0] == player and board[1] == player and board[2] == player) or
            (board[3] == player and board[4] == player and board[5] == player) or
            (board[6] == player and board[7] == player and board[8] == player) or
            (board[0] == player and board[3] == player and board[6] == player) or
            (board[1] == player and board[4] == player and board[7] == player) or
            (board[2] == player and board[5] == player and board[8] == player) or
            (board[0] == player and board[4] == player and board[8] == player) or
            (board[2] == player and board[4] == player and board[6] == player))

# define a function to check if the board is full
def check_full():
    return " " not in board

# loop until the game is over
while True:
    # prompt the user to choose X or O
    player = input("Do you want to be X or O? ")

    # assign the computer the opposite symbol
    computer = "X" if player == "O" else "O"

    # print the starting board
    print_board()

    # loop until the game is over
    while True:
        # prompt the user to make a move
        move = input("Enter your move (0-8): ")

        # check if the move is valid
        if board[int(move)] == " ":
            board[int(move)] = player
            print_board()

            # check if the player has won
            if check_win(player):
                print("Congratulations, you have won!")
                break

            # check if the board is full
            if check_full():
                print("The game is a tie.")
                break

            # let the computer make a move
            computer_move = random.choice([i for i in range(9) if board[i] == " "])
            board[computer_move] = computer
            print("Computer has made a move.")
            print_board()

            # check if the computer has won
            if check_win(computer):
                print("Sorry, the computer has won.")
                break

        # if the move is invalid, prompt the user to try again
        else:
            print("Invalid move, please try again.")

    # prompt the user to play again or quit
    play_again = input("Do you want to play again? (Y/N) ")
    if play_again.lower() != "y":
        break
```
