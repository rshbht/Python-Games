```python
import random

# Define the deck of cards
deck = []
colors = ['Red', 'Green', 'Blue', 'Yellow']
special_cards = ['Skip', 'Reverse', 'Draw Two']
wild_cards = ['Wild', 'Wild Draw Four']

for color in colors:
    for i in range(0, 10):
        deck.append(f'{color} {i}')
    for special_card in special_cards:
        deck.append(f'{color} {special_card}')

for wild_card in wild_cards:
    for i in range(0, 4):
        deck.append(wild_card)

# Shuffle the deck
random.shuffle(deck)

# Define the players
players = []
num_players = int(input('How many players? '))
for i in range(1, num_players+1):
    player_name = input(f'Enter name for player {i}: ')
    players.append({'name': player_name, 'hand': []})

# Deal the cards
num_cards_to_deal = 7
for player in players:
    player['hand'] = deck[0:num_cards_to_deal]
    del deck[0:num_cards_to_deal]

# Define the starting card
starting_card = deck.pop()

# Define the game loop
current_player_index = 0
direction = 1
is_game_over = False
while not is_game_over:
    current_player = players[current_player_index]
    print(f"\n\n{current_player['name']}'s turn")
    print(f"Current card: {starting_card}")
    print(f"Your hand: {current_player['hand']}")
    playable_cards = [card for card in current_player['hand'] if card.startswith(starting_card.split(' ')[0]) or card.split(' ')[0] in ['Wild', 'Wild Draw Four']]
    if not playable_cards:
        current_player['hand'].append(deck.pop())
        print(f"{current_player['name']} draws a card")
    else:
        print(f"Playable cards: {playable_cards}")
        chosen_card = input("Enter card to play or 'draw' to draw a card: ")
        if chosen_card == 'draw':
            current_player['hand'].append(deck.pop())
            print(f"{current_player['name']} draws a card")
        else:
            current_player['hand'].remove(chosen_card)
            starting_card = chosen_card
            if starting_card.startswith('Wild Draw Four'):
                next_player_index = (current_player_index + direction) % len(players)
                players[next_player_index]['hand'].extend(deck[0:4])
                del deck[0:4]
            elif starting_card.startswith('Draw Two'):
                next_player_index = (current_player_index + direction) % len(players)
                players[next_player_index]['hand'].extend(deck[0:2])
                del deck[0:2]
            elif starting_card == 'Reverse':
                direction = -direction
            elif starting_card == 'Skip':
                current_player_index = (current_player_index + direction) % len(players)
            elif starting_card.startswith('Wild'):
                color = input("Choose a color (Red, Green, Blue, or Yellow): ")
                starting_card = f"{color} Wild"
            current_player_index = (current_player_index + direction) % len(players)
    if len(current_player['hand']) == 0:
        print(f"{current_player['name']} wins!")
        is_game_over = True
```
